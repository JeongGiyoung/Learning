## ## [타자 데이터]

```python
def getAllDataHitter(year = 2022)  :
  """
  # year = 검색할 연도
  """

  driver.get("https://www.koreabaseball.com/Record/Team/Hitter/Basic1.aspx")

  dropdown = Select(driver.find_element_by_id('cphContents_cphContents_cphContents_ddlSeason_ddlSeason'))
  dropdown.select_by_value(f"{year}")

  sleep(10)

  html = driver.page_source

  # 파서
  soup = BeautifulSoup(html, "html.parser")

  # 컬럼명 추출
  columns_all = list()

  for i in soup.select("th")[1:] :
    columns_all.append(i.text)

  # 팀명 추출
  team = list()

  for i_team_list in range(1,len(soup.select("tr"))-1) :
    team_xpath = driver.find_element_by_xpath(f"//*[@id='cphContents_cphContents_cphContents_udpContent']/div[2]/table/tbody/tr[{i_team_list}]/td[2]")
    team.append(team_xpath.text)

  # 추출할 td 목록
  td_list_1 = ["team"
              , "HRA_RT"
              , "GAME_CN"
              , "PA_CN"
              , "AB_CN"
              , "RUN_CN"
              , "HIT_CN"
              , "H2_CN"
              , "H3_CN"
              , "HR_CN"
              , "TB_CN"
              , "RBI_CN"
              , "SH_CN"
              , "SF_CN"
              ]
    
  # 각 수치 변수에 넣기
  HRA_RT = list()
  GAME_CN = list()
  PA_CN = list()
  AB_CN = list()
  RUN_CN = list()
  HIT_CN = list()
  H2_CN = list()
  H3_CN = list()
  HR_CN = list()
  TB_CN = list()
  RBI_CN = list()
  SH_CN = list()
  SF_CN = list()

  for idx in td_list_1 :
    for i in range(len(soup.select(f"[data-id='{idx}']"))) :
      if idx == "HRA_RT" :
        HRA_RT.append(soup.select(f"[data-id='{idx}']")[i].text)
      elif idx == "GAME_CN" :
        GAME_CN.append(soup.select(f"[data-id='{idx}']")[i].text)
      elif idx == "PA_CN" :
        PA_CN.append(soup.select(f"[data-id='{idx}']")[i].text)
      elif idx == "AB_CN" :
        AB_CN.append(soup.select(f"[data-id='{idx}']")[i].text)
      elif idx == "RUN_CN" :
        RUN_CN.append(soup.select(f"[data-id='{idx}']")[i].text)
      elif idx == "HIT_CN" :
        HIT_CN.append(soup.select(f"[data-id='{idx}']")[i].text)
      elif idx == "H2_CN" :
        H2_CN.append(soup.select(f"[data-id='{idx}']")[i].text) 
      elif idx == "H3_CN" :
        H3_CN.append(soup.select(f"[data-id='{idx}']")[i].text)  
      elif idx == "HR_CN" :
        HR_CN.append(soup.select(f"[data-id='{idx}']")[i].text)  
      elif idx == "TB_CN" :
        TB_CN.append(soup.select(f"[data-id='{idx}']")[i].text)
      elif idx == "RBI_CN" :
        RBI_CN.append(soup.select(f"[data-id='{idx}']")[i].text)
      elif idx == "SH_CN" :
        SH_CN.append(soup.select(f"[data-id='{idx}']")[i].text)
      elif idx == "SF_CN" :
        SF_CN.append(soup.select(f"[data-id='{idx}']")[i].text)

  # 데이터 프레임으로 만들기
  all_data_hitter_1 = pd.DataFrame(data = list(zip(team
                                          ,HRA_RT
                                          ,GAME_CN
                                          ,PA_CN
                                          ,AB_CN
                                          ,RUN_CN
                                          ,HIT_CN
                                          ,H2_CN
                                          ,H3_CN
                                          ,HR_CN
                                          ,TB_CN
                                          ,RBI_CN
                                          ,SH_CN
                                          ,SF_CN))
                        , columns = columns_all
                        ).set_index('팀명')

  # 2페이지 
  driver.get("https://www.koreabaseball.com/Record/Team/Hitter/Basic2.aspx")

  dropdown = Select(driver.find_element_by_id('cphContents_cphContents_cphContents_ddlSeason_ddlSeason'))
  dropdown.select_by_value(f"{year}")

  sleep(10)

  html = driver.page_source
  soup = BeautifulSoup(html, "html.parser")

  td_list_2 = ["team"
              , "HRA_RT"
              , "BB_CN"
              , "IB_CN"
              , "HP_CN"
              , "KK_CN"
              , "GD_CN"
              , "SLG_RT"
              , "OBP_RT"
              , "OPS_RT"
              , "MH_HITTER_CN"
              , "SP_HRA_RT"
              , "PH_HRA_RT"
              ]

  HRA_RT =list()
  BB_CN= list()
  IB_CN= list()
  HP_CN= list()
  KK_CN= list()
  GD_CN= list()
  SLG_RT= list()
  OBP_RT= list()
  OPS_RT= list()
  MH_HITTER_CN= list()
  SP_HRA_RT= list()
  PH_HRA_RT= list()

  for idx in td_list_2 :
    for i in range(len(soup.select(f"[data-id='{idx}']"))) :
      if idx == "BB_CN" :
        BB_CN.append(soup.select(f"[data-id='{idx}']")[i].text)
      elif idx == "HRA_RT" :
        HRA_RT.append(soup.select(f"[data-id='{idx}']")[i].text)    
      elif idx == "IB_CN" :
        IB_CN.append(soup.select(f"[data-id='{idx}']")[i].text)
      elif idx == "HP_CN" :
        HP_CN.append(soup.select(f"[data-id='{idx}']")[i].text)
      elif idx == "KK_CN" :
        KK_CN.append(soup.select(f"[data-id='{idx}']")[i].text)
      elif idx == "GD_CN" :
        GD_CN.append(soup.select(f"[data-id='{idx}']")[i].text)
      elif idx == "SLG_RT" :
        SLG_RT.append(soup.select(f"[data-id='{idx}']")[i].text)
      elif idx == "OBP_RT" :
        OBP_RT.append(soup.select(f"[data-id='{idx}']")[i].text)
      elif idx == "OPS_RT" :
        OPS_RT.append(soup.select(f"[data-id='{idx}']")[i].text)
      elif idx == "MH_HITTER_CN" :
        MH_HITTER_CN.append(soup.select(f"[data-id='{idx}']")[i].text)
      elif idx == "SP_HRA_RT" :
        SP_HRA_RT.append(soup.select(f"[data-id='{idx}']")[i].text)
      elif idx == "PH_HRA_RT" :
        PH_HRA_RT.append(soup.select(f"[data-id='{idx}']")[i].text)

  # 컬럼명 추출
  columns_all = list()

  for i in soup.select("th")[1:] :
    columns_all.append(i.text)
  
  all_data_hitter_2 = pd.DataFrame(data = list(zip(team
                                          , HRA_RT
                                          , BB_CN
                                          ,IB_CN
                                          ,HP_CN
                                          ,KK_CN
                                          ,GD_CN
                                          ,SLG_RT
                                          ,OBP_RT
                                          ,OPS_RT
                                          ,MH_HITTER_CN
                                          ,SP_HRA_RT
                                          ,PH_HRA_RT
                                        )
                                      )
                        , columns = columns_all
                        ).set_index('팀명')


  return pd.concat([all_data_hitter_1,all_data_hitter_2],axis=1)
```
